```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate positional encodings
def generate_positional_encodings(seq_len, embedding_dim):
    position = np.arange(seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, embedding_dim, 2) * -(np.log(10000.0) / embedding_dim))
    pos_enc = np.zeros((seq_len, embedding_dim))
    pos_enc[:, 0::2] = np.sin(position * div_term)
    pos_enc[:, 1::2] = np.cos(position * div_term)
    return pos_enc

# Parameters
sequence_length = 100
embedding_dimension = 32

# Generate positional encodings
positional_encodings = generate_positional_encodings(sequence_length, embedding_dimension)

# Plot heatmap of positional encodings
plt.figure(figsize=(10, 6))
sns.heatmap(positional_encodings, cmap="viridis", cbar=True)
plt.title("Positional Encoding Heatmap")
plt.xlabel("Embedding Dimension")
plt.ylabel("Sequence Position")
plt.show()

# Plot individual sine and cosine functions for different embedding dimensions
plt.figure(figsize=(12, 6))
for i in range(0, embedding_dimension, 4):  # Plot every 4th dimension for clarity
    plt.plot(positional_encodings[:, i], label=f"Sine Dim {i}")
    plt.plot(positional_encodings[:, i + 1], label=f"Cosine Dim {i + 1}")

plt.title("Sine and Cosine Functions for Positional Encoding")
plt.xlabel("Sequence Position")
plt.ylabel("Value")
plt.legend()
plt.show()
```
